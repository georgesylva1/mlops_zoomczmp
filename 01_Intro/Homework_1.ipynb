```python
df_1 = pd.read_parquet("https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-01.parquet")
df_2 = pd.read_parquet("https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-02.parquet")
```

```python
# Check for missing values for Jan
print(df_1.isnull().sum())
```

```python
# Calculate trip duration in minutes
df_1['duration_minutes'] = (df_1['tpep_dropoff_datetime'] - df_1['tpep_pickup_datetime']).dt.total_seconds() / 60

# Calculate standard deviation of trip duration
duration_std = df_1['duration_minutes'].std()

print(f"Standard deviation of trip duration in January: {duration_std:.2f} minutes")
```

```python
# Count total original records
total_records = len(df_1)

# Filter records (1 ≤ duration ≤ 60 minutes)
df_1_filtered = df_1[(df_1['duration_minutes'] >= 1) & (df_1['duration_minutes'] <= 60)]

# Count remaining records
remaining_records = len(df_1_filtered)

# Calculate fraction remaining
fraction_remaining = remaining_records / total_records

print(f"Original records: {total_records:,}")
print(f"Records after filtering: {remaining_records:,}")
print(f"Fraction remaining: {fraction_remaining:.2%}")
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
df_1['duration_minutes'].plot(kind='box')
plt.title('Original Durations')

plt.subplot(1, 2, 2)
df_1_filtered['duration_minutes'].plot(kind='box')
plt.title('Filtered Durations (1-60 mins)')
plt.show()
```

```python
from sklearn.feature_extraction import DictVectorizer

# Convert the filtered dataframe to a list of dictionaries with string IDs
location_dicts = df_1_filtered[['PULocationID', 'DOLocationID']].astype(str).to_dict('records')

# Initialize and fit the dictionary vectorizer
dv = DictVectorizer(sparse=True)  # Set sparse=True for large datasets to save memory
dv.fit(location_dicts)

# Now we can safely check feature names
print(f"First 10 feature names: {dv.feature_names_[:10]}")
```

```python
# Transform the data to a feature matrix
feature_matrix = dv.transform(location_dicts)

# Get the dimensionality
num_columns = feature_matrix.shape[1]
print(f"Dimensionality of the feature matrix: {num_columns} columns")
print(f"Matrix format: {feature_matrix.format}")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Prepare the target variable (duration in minutes)
y = df_1_filtered['duration_minutes'].values

# Train the linear regression model
model = LinearRegression()
model.fit(feature_matrix, y)

# Make predictions on training data
y_pred = model.predict(feature_matrix)

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y, y_pred))
print(f"Training RMSE: {rmse:.2f} minutes")

# Optional: Print R-squared score
r2 = model.score(feature_matrix, y)
print(f"R-squared: {r2:.4f}")
```

```python
# For very large datasets, use SGDRegressor instead
from sklearn.linear_model import SGDRegressor
sgd_model = SGDRegressor(loss='squared_error', penalty=None)
sgd_model.fit(feature_matrix, y)
y_pred_sgd = sgd_model.predict(feature_matrix)
print(f"SGD RMSE: {np.sqrt(mean_squared_error(y, y_pred_sgd)):.2f} minutes")
```

```python
# Prepare February data (df_2) the same way we did for January

# Filter February trips to 1-60 minute durations
df_2['duration_minutes'] = (df_2['tpep_dropoff_datetime'] - df_2['tpep_pickup_datetime']).dt.total_seconds() / 60
df_2_filtered = df_2[(df_2['duration_minutes'] >= 1) & (df_2['duration_minutes'] <= 60)]
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
df_2['duration_minutes'].plot(kind='box')
plt.title('Original Durations')

plt.subplot(1, 2, 2)
df_2_filtered['duration_minutes'].plot(kind='box')
plt.title('Filtered Durations (1-60 mins)')
plt.show()
```

```python
from scipy.sparse import vstack

# Process validation data in batches
batch_size = 5000
val_dicts = df_2_filtered[['PULocationID', 'DOLocationID']].astype(str).to_dict('records')

X_val_batches = []
for i in range(0, len(val_dicts), batch_size):
    batch = val_dicts[i:i + batch_size]
    X_val_batches.append(dv.transform(batch))  # Use the pre-fitted vectorizer

# Combine all batches into one sparse matrix
X_val = vstack(X_val_batches)
y_val = df_2_filtered['duration_minutes'].values

# Predict in batches
y_val_pred = []
predict_batch_size = 5000
for i in range(0, X_val.shape[0], predict_batch_size):
    batch = X_val[i:i + predict_batch_size]
    y_val_pred.extend(model.predict(batch))
    
y_val_pred = np.array(y_val_pred)

# Calculate RMSE
val_rmse = np.sqrt(mean_squared_error(y_val, y_val_pred))
print(f"Validation RMSE: {val_rmse:.2f} minutes")
```
